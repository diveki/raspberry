# Díszítsd fel a beszéded

Eddigiekben már kaptunk némi bepillantást abba, hogy egy hatásos előadáshoz milyen fontos a beszéd megtervezése. Először a beszéd szerkezetét tárgyaltuk, bevezető, fő rész és konklúzió, majd áttértünk egy egyértelmű üzenet megfogalmazására ami köré az egész beszédet építjük. 

Ha megvannak az üzeneteink és azok kapcsolatai, akkor gondolkodhatunk azon, hogyan öltöztetjük fel, hogyan díszítjük ki a beszédünket. Ebben segíthetnek a retorikai eszközök.

Ezeket az eszközöket legtöbbször észre sem vesszük, pedig néha még a hétköznapi életünkben is használjuk. Olyan kommunikációs eszközök, olyan titkos minták, amelyek hatást váltanak ki a közönségből. Ismeretükben tervezhetjük alapból a beszédet velük, de akár útólag az elkészült beszédbe beépítve erősíthetjük annak üzenetét. Ezen eszközök ismerete nem csak megnyerőbb előadóvá tesz minket hanem más beszédekben is segít felismerni ezeket az elemeket.

A retorikai eszközöket mi most két részre osztjuk (habár más fajta csoportosítások is léteznek):

1) Szóképek - célja, hogy a verbális vizualizáción kereszül az érzelmekre, a memorizálásra hassunk ilyenek pl. megszemélyesítés, metafóra, hasonlat stb. 

2) Alakzatok - az egyes üzenetek, gondolatok megerősítésére szolgálnak. Ilyen pl. az ellentét, a túlzás, a szóhalmozás.

**A beszédekben a jellemző sorrendjük, hogy képi eszközökkel kezdünk (érzelmi hangolás), segítünk a közönségnek látni, érezni (ízlelés, érintés, szaglás), vízionálni azt amiről beszélünk majd ha eljutottunk az üzenetekhez, akkor használunk több alakzatok, hogy az üzenet leülepedjen a hallgatóságban.**

## Érzelmi hangolás

A beszéd készítése sose legyen öncélú. A beszédet másoknak adjuk elő, ezért mindig azt kell szem előtt tartani, hogy mit nyer a közönség az előadásunk meghallgatásával. Elsődlegesen az üzenetet kellene megkapnia a közönségnek, de hogy még hatásosabb legyen az átadás érzelmi töltetet is lehet adni nekik. 

Ha a beszédünket úgy írjuk meg, hogy a közönség bele tudja magát képzelni a mondatainkba, képzeletben látja, ízleli, szagolja, tapintja, hallja vagy átérzi mindazt amiről beszélünk, akkor egy maradandó élménnyel gazdagszik és biztos emlékezni fog ránk vagy a beszédre. Ehhez olyan leíró nyelvezetet kell használnunk, ami lényegében egy színházba tereli a közönséget. 

Ha szeretnénk, hogy egy bizonyos kifejezésre, gondolatra, fogalomra koncentráljon a közönség, akkor azt nyomatékosíthatjuk ismétléssel, rövid mondatokban és mindig kicsit másképp megfogalmazva (de csak max háromszori ismétléssel). 

Az eszközök használatánal nagyon fontos az időzítés is. Ha nincs jól előkészítve, felvezetve a terep, akkor ezek a retorikai eszközök veszítenek a hatékonyságukból. 

# Példák a retorikai eszközökre

## Megszemélyesítés

**Mi ez**: elvont dolgokat, élettelen tárgyakat vagy természeti jelenségeket, illetve növényeket vagy állatokat az emberekre jellemző érzésekkel és tulajdonságokkal ruház fel.

**Mire/mikor használjuk**: emberibbé, ember közelivé teszi a beszéd tárgyát, amivel a közönség könnyebben azonosul.

**Példák**: a pillanat nagysága súgta meg, hogy mikor mi a teendőnk; A csillagok játékosan táncoltak a holdfényben úszó égen; kifut a leves; siet az óra; nem akar elindulni az autó; S szelíd fátylával száll a csendes est stb.

## Metafóra 

**Mi ez**: A metafóra egy ismeretlen tárgyat ruház fel egy ismert tárgy tulajdonságával. Két dolognak az összehasonlítása, azok azonosítása, kapcsolatba hozása, amely kapcsolat lehet külső, belső vagy hangulati hasonlóság.

**Mire használjuk**: Használható bonyolult összefüggések elmagyarázására, emlékezetessé tételére.

**Példák**: Színház az egész világ. És színész benne minden férfi és nő.; elhasal a vizsgán; A szerelem sötét verem; Kikelet a lány;

## Hasonlat

**Mi ez**: Két dolog vagy azok tulajdonságainak az összehasonlítása az “olyan, mint” formulával.

**Mire használjuk**:  Láthatóbbá, tapinthatóbbá, valószerűbbé teszi a gondolatot, amelyet ki akarunk fejezni. A hasonlatnak köszönhetően a hasonlított fogalom elképzelhetőbbé, árnyaltabbá válik. Összefüggések gyors szemléltetésére és gyors értelmezésére használható.

**Példák**: Józsi bácsi vén, mint az országút; Aranyos, mint egy kiscica; arany, mint a méz; Kicsit nagyobb, mint egy futballpálya;

## Túlzás / nagyítás

**Mi ez**: Valaminek a kifejezése eltúlozva azért, hogy komolyabbnak, fontosabbnak vagy érdekesebbnek tűnjön. Lehet egy humor eszköz is. 

**Mire használjuk**: Egy adott témára irányítja a figyelmet. Körülmények szándékos eltúlzása nyomatékosítás vagy hatás növelése céljából. Vagy összehasonlításra is használható olyan módon, hogy az egyik felvetést arányosan, a másikat pedig eltúlozva mutatjuk be. A létrehozott leírásnak meglepetést vagy örömet okozhat a közönségben az összehasonlítás vagy a leírás által ábrázolt kép miatt.

**Példák**: Annyira nézik a tévét, hogy kipottyan a szemük; A táska tonnás súlyu volt; Elég szőke volt ahhoz, hogy a püspök lyukat rúgjon egy ólomüveg ablakon keresztül; A tó olyan csendes és tiszta volt, hogy át lehetett látni rajta a Föld közepéig.

## Retorikai kérdés / szónoki kérdés / költői kérdés

**Mi ez**: Ez egy olyan kérdés, amit a beszélő a közönségnek tesz fel és ugyan nem vár választ rá, mégis a közönséget a téma színhelyére repíti és a nyilvánvaló válasszal egyesíti. Célja, hogy a közönség magában keressen választ, gondolkodjon el, de ne mondja ki hangosan. A költői kérdés gyakran sugalmazza a válaszát: a kérdés jellege vagy megfogalmazása miatt a válasz triviális

**Mire használjuk**: Interakcióra a beszéd közben, a közönség bevonására, egy fontos gondolat kiemelésére azzal a céllal, hogy érvényre juttasson egy gondolatot. Ha lankad a figyelem egy kérdés mindig figyelemfelkeltő hatással bír.

**Példák**: Nem dolgozunk mindannyian túl sokat?; Akarjátok látni még egyszer?; Szeretnétek kimaradni a tortaevésből?; Lehet játékos projekteket készíteni a Raspberry Pi-al?

## Alliteráció

**Mi ez**: A szókezdő mássalhangzók ismétlődése. 

**Mire használjuk**: Hangzásukkal felhívják a figyelmet az általuk közölt tartalomra.

**Példák**: Csenevész csellengő csepűrágó csemete cselesen cseberből cseresznyét csemegéz; hitemet hirdeti híven; Hallgatják halkan a harmonikát; S hallgatom a fák lehulló levelének lágy neszét;

## Antitézis / Ellentét

**Mi ez**: Két gondolat közti kapcsolat kontrasztozása, ellentétbe helyezése. A konfliktus minden jó mese, történet és üzleti előadás egyik legfőbb eszköze. Az ellentétet, mint retorikai eszközt egy mondaton belül értjük.

**Mire használjuk**: Mivel két dolog van szembe állítva, így a mondandót jobban kifejezi, mint a felsorolás, jobban képes megértést vagy empátiát kiváltani. Segít a pro-t és kontrát megvilágítani és dönteni két lehetőség között.

**Példák**: Kis lépés egy embernek, de hatalmas ugrás az emberiségnek.; Jó hal voltam rossz tóban.; én szeretem önt – ön nem szeret engem.; Inkább ma száraz barna kenyeret, Mint holnap lágy fehér cipót.


## Aporia

**Mi ez**: Egy megjátszott kijelentés vagy kétely kifejezése legtöbbször kérdés formájában. Úgy tűnik, mintha a szónok csak hangosan gondolkodna, de közben ezzel a közös gondolkodással vezeti a közönséget.

**Mire használjuk**: Egy kérdés közönség elgondolkodtatására vagy megdolgoztatására, esetleg ismert érv árnyalására. A szónok bizonytalansága a közönséget arra sarkallja, hogy segítsen neki a válasz megadásával. Segít erősíteni a kapcsolatot a szónok és a közönség között.

**Példák**: Hogyan fogjuk ezt kommunikálni?; Mit fogunk tenni?

## Kibővítés / Felerősítés

**Mi ez**: Egyes szavak vagy kifejezések ismétlése, s közben egyre több részlettel vannak kiegészítve.

**Mire használjuk**: Ha valaminek a jelentősségét szeretnénk hangsúlyozni, akkor ismétlünk és minden egyes ismétlést tovább bővítve fogalmazzuk meg.

**Példák**: Egy házban lakom. Egy kertes házban. Egy kertes házban a falu közepén.; Kaland. Az életem egy kaland. Az életem egy kaland, Tom Sawyer életéhez hasonlóan.

## Anafora / előismétlés

**Mi ez**:  Egy vagy több szó ismétlődése két vagy több egymást követő tagmondat, mondat elején. A szóismétlés speciális módszere.

**Mire használjuk**: Felerősíti egy gondolat, egy szó, egy fogalom értékét a beszédben, így fókuszálva a közönség figyelmét az adott szövegre.

**Példák**: A szerelem a kérdés. A szerelem a válasz. A szerelem a minden.; micsoda már a délután, micsoda már; 

## Antimetabola

**Mi ez**: Szavak vagy kifejezések megismétlése fordított sorrendben

**Mire használjuk**: Arra szolgál, hogy egy gondolatot, kifejezést elültessünk a közönség fejében.

**Példák**: Amikor kétségbe vonod az erőd, erőt adsz a kétségednek.; Jaj, mily sekély a mélység és mily mély a sekélység.; Egyél, hogy élj, ne azért élj, hogy egyél.; Ne azt kérdezd, hogy mit tud érted tenni a hazád, hanem azt, hogy te mit tudsz tenni a hazádért;


## Fokozás

**Mi ez**: A szavak, gondolatok és mondatok rendezése emelkedő fontossági sorrendbe. A jó, jobb, legjobb struktúra. A reklámokból a Gut, Besser, Gösser lehet ismerős.

**Mire használjuk**: A fokozás várakozást és izgalmat teremt a közönségben. Ahogy haladunk a fokozásban a közönség úgy kerül izgatottabb állapotba. Kiválóan használható meggyőzésre.

**Példák**: Barcelonában jó az étel. Ami még jobb, az a Mediterrán hangulat. A legjobb dolog viszont az éghajlat.; Hass, alkoss, gyarapíts!; 

## Utóismétlés

**Mi ez**: Egymást követő kifejezések, tagmondatok vagy mondatok végén előforduló ismétlés.

**Mire használjuk**: Felerősíti egy gondolat, egy szó, egy fogalom értékét a beszédben, így fókuszálva a közönség figyelmét az adott szövegre.

**Példák**: mikor gyermek voltam, úgy szóltam, mint gyermek, úgy éreztem, mint gyermek, úgy gondolkoztam, mint gyermek.; 




# Konklúzió

Mindig tartsuk szem előtt, hogy az előadást nem magunkért, hanem a közönségért tartjuk, így az ő érdekeik a legfontosabbak. Igyekezzünk biztosítani őket, hogy minél jobb érzelmekkel és gondolatokkal hagyják el az előadó termet és magukkal vigyék a nekik szánt üzenetet. 

A retorikai eszközök csak segédeszközök. Azt a célt szolgálják, hogy az üzenet amit át szeretnénk adni, hatásosabban és mélyebben érje el a közönséget.
